```python
import cmath

# Given impedances 
Z1 = 5 + 3j      
Z2 = 2 - 4j      
Z3 = 1 + 6j      
```

```python
# i. Total impedance 
Z_total = Z1 + Z2 + Z3
```

```python
# ii. Real and imaginary parts
real_part = Z_total.real      
imag_part = Z_total.imag      
```

```python
# iii. Magnitude and phase angle
magnitude = abs(Z_total)                
phase_angle = cmath.phase(Z_total)       
```

```python
# Print results
print("Total Impedance Z =", Z_total)
print("Real part (Resistance) =", real_part, "Ω")
print("Imaginary part (Reactance) =", imag_part, "Ω")
print("Magnitude |Z| =", magnitude, "Ω")
print("Phase Angle =", phase_angle, "radians")
```
