Task 1: Create and Write Energy Data to a CSV File

```python
import csv
```

```python
# Create and write initial energy data
with open("data.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Date", "Energy consumed in kWh"])
    writer.writerow(["2025-01-01", 120.5])
    writer.writerow(["2025-01-02", 118.2])
    writer.writerow(["2025-01-03", 125.0])
```

```python
print("data.csv created and data written successfully.")
```

Task 2: Read and Display CSV File Data

```python
import csv

print("\nEnergy Consumption Records:\n")

with open("data.csv", mode="r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(f"{row[0]:<15} {row[1]}")
```

Task 3: Append New Energy Records

```python
import csv
```

```python
# Append new energy consumption records

with open("data.csv", mode="a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["2025-01-04", 122.7])
    writer.writerow(["2025-01-05", 119.9])
```

```python
print("New energy records appended successfully.")
```
