```python
# 1. List of student names

students = ["Ali", "Usman", "Hassan", "Kamran"]
```

```python
# 2. Tuple of student roll numbers (immutable)

roll_numbers = (101, 102, 103, 104)
```

```python
# 3. Nested list of grades (3 subjects each)

grades = [
    [85, 90, 78],   # Ali
    [88, 92, 80],   # Usman
    [75, 85, 82],   # Hassan
    [90, 88, 91]    # Kamran
]
```

```python
# 4. Update Ali’s second grade to 95

grades[0][1] = 95
```

```python
# 5. Add a new grade for Hassan (4th subject)

grades[2].append(89)
```

```python
# 6. Attempt to change Kamran’s roll number

try:
    roll_numbers[3] = 110
except TypeError as e:
    print("Error:", e)
```

```python
# 7. Print student details

print("\nStudent Details:")
for i in range(len(students)):
    print(f"Name: {students[i]}, Roll No: {roll_numbers[i]}, Grades: {grades[i]}")
```
