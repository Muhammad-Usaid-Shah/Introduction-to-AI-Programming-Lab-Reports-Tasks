```python
# I. Create a dictionary with 5 fields

reading1 = {
    "solar_power": 250,        # watts
    "panel_current": 8.5,      # amps
    "battery_voltage": 12.6,   # volts
    "load_status": "ON",       # ON / OFF
    "temperature": 35          # degree Celsius
}
```

```python
# II. Print required values

print("Solar Power:", reading1["solar_power"])
print("Battery Voltage:", reading1["battery_voltage"])
print("Load Status:", reading1["load_status"])
```

```python
# III. Print number of fields in dictionary

print("Number of fields:", len(reading1))
```

```python
# IV. Remove the 'panel_current' field

reading1.pop("panel_current")
```

```python
# V. Get Keys, Values, and Items
print("\nAll Keys:")
print(reading1.keys())

print("\nAll Values:")
print(reading1.values())

print("\nAll Key-Value Pairs:")
print(reading1.items())
```
